```python
#https://towardsdatascience.com/how-to-convert-json-into-a-pandas-dataframe-100b2ae1e0d8
#https://github.com/BindiChen/machine-learning/blob/main/data-analysis/027-pandas-convert-json/pandas-convert-json.ipynb
import pandas as pd
original_article_list = pd.read_csv('/content/drive/MyDrive/GitHub/PubMed_pipeline/LAG/lag_articles_list.csv')
abstracts = pd.read_json("/content/drive/MyDrive/GitHub/PubMed_pipeline/LAG/annotated_abstracts.json", lines=True)
fulltext = pd.read_json("/content/drive/MyDrive/GitHub/PubMed_pipeline/LAG/annotated_fulltext.json", lines=True)
```

```python
abstracts["PMID"] = abstracts["pmid"]
fulltext["PMID"] = fulltext["pmid"]
abstracts["PMCID_abs"] = abstracts["pmcid"]
fulltext["PMCID_ft"] = fulltext["pmcid"]
abstracts["abstract_passages"] = abstracts["passages"]
fulltext["fulltext_passages"] = fulltext["passages"]
abstracts["abstract_accessions"] = abstracts["accessions"]
fulltext["fulltext_accessions"] = fulltext["accessions"]
abs_subset = abstracts[["PMID", "PMCID_abs", "abstract_passages", "abstract_accessions"]]
fulltext_subset = fulltext[["PMID", "PMCID_ft", "fulltext_passages", "fulltext_accessions"]]
```

```python
fulltext_subset.head(1)
```

```python
article_list = pd.merge(original_article_list, abs_subset, on='PMID', how='left')
article_list1 = pd.merge(article_list, fulltext_subset, on='PMID', how='left')
article_list1
```

```python
article_list1.to_csv('/content/drive/MyDrive/GitHub/PubMed_pipeline/LAG/lag_articles_list_with_retrieved_data.csv', index=False)
```
